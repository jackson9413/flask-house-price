```python
# draw a heatmap to show the correlation between other columns and the price column
# exclude id, date, zip code, latitude, longitude
# sort by correlation with price
# show half of the heatmap
plt.figure(figsize=(20,20))
sns.heatmap(df.drop(['id', 'date', 'zipcode', 'lat', 'long'], axis = 1).corr(), annot=True, cmap='coolwarm', center=0)
```

```python
# columns whose correlation with price larger than 0.3
df.drop(['id', 'date', 'zipcode', 'lat', 'long'], axis = 1).corr().price[df.drop(['id', 'date', 'zipcode', 'lat', 'long'], axis = 1).corr().price > 0.3]
```

```python
# draw a heat map to show the correlation between the columns whose correlation with price larger than 0.3
# SHOW THIS PLOT
plt.figure(figsize=(20,20))
sns.heatmap(df[["price", "bedrooms", "bathrooms", "sqft_living", "sqft_above", "sqft_basement"]].corr(), annot=True, cmap='coolwarm', center=0)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.drop(['id', 'date', 'zipcode', 'lat', 'long'], axis = 1).corr().loc[['price', 'sqft_living', 'grade', 'sqft_above', 'sqft_living15', 'bathrooms', 'view', 'sqft_basement', 'bedrooms']], 
            annot=True, cmap='coolwarm', center=0)
```

```python
# show the relationship between price and sqft_living, sqft_above, sqft_basement 
# SHOW THIS PLOT
plt.figure(figsize=(10,10))
sns.scatterplot(x = 'sqft_living', y = 'price', data = df)
plt.figure(figsize=(10,10))
sns.scatterplot(x = 'sqft_above', y = 'price', data = df)
plt.figure(figsize=(10,10))
sns.scatterplot(x = 'sqft_basement', y = 'price', data = df)
```

```python
# zip code unique count
df['zipcode'].nunique()
```

```python
# draw a plot to show the relationship between zip code and price
plt.figure(figsize=(20,10))
sns.boxplot(x = 'zipcode', y = 'price', data = df)
```

```python
# zip code with the top 10 average price
df.groupby('zipcode').price.mean().nlargest(10)
```

```python
# draw a box plot to show number of bedrooms and price
plt.figure(figsize=(20,10))
sns.boxplot(x = 'bedrooms', y = 'price', data = df)
```

```python
# group by number of bedrooms and count the number of houses
# sort by number of houses in descending order 
df.groupby('bedrooms').size().sort_values(ascending = False)
```

Most of the houses have one to 6 bedrooms.

```python
# draw a box plot to shown houses with 1 to 6 bedrooms and price
plt.figure(figsize=(20,10))
sns.boxplot(x = 'bedrooms', y = 'price', data = df[df['bedrooms'] <= 6])
```

```python
# draw a box plot to show number of bathrooms and price
plt.figure(figsize=(20,10))
sns.boxplot(x = 'bathrooms', y = 'price', data = df)
```

```python
# group by number of bathrooms and count the number of houses
# sort by number of houses in descending order
df.groupby('bathrooms').size().sort_values(ascending = False)
```

Most of the houses have 1 to 4 bathrooms.

```python
# draw a box plot to shown houses with 1 to 4 bathrooms and price
# SHOW THIS PLOT
plt.figure(figsize=(20,10))
sns.boxplot(x = 'bathrooms', y = 'price', data = df[df['bathrooms'] <= 4])
```

```python
# draw a box plot to show the relationship between grade and price
# display the underlying data points
plt.figure(figsize=(20,10))
sns.boxplot(x = 'grade', y = 'price', data = df)
#sns.swarmplot(x = 'grade', y = 'price', data = df, color = 'k', alpha = 0.5)
```

```python
# group by grade and count the number of houses
# sort by number of houses in descending order
df.groupby('grade').size().sort_values(ascending = False)
```
